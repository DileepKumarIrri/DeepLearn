```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

# Training data for XOR gate
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[0],[1],[1],[0]], "float32")

# Create a sequential model
model = Sequential()

# Add the input layer
model.add(Dense(32, input_dim=2, activation='relu')) 
# 32 neurons in the input layer with ReLU activation function, 32 because it doesn't satisfy all the gates
# 32 is the minimum no. of nodes required to satisfy all the gates with 200 epochs

# Add the output layer
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])

# Train the model
model.fit(training_data, target_data, epochs=200, verbose=2)

# Evaluate the model
print(model.evaluate(training_data, target_data))

# Predict the result for XOR gate
print(model.predict(training_data).round())
```
