Without a dataset

```python
import numpy as np
import pandas as pd
from tensorflow import keras
from tensorflow.keras import layers
```

```python
x = 2 * np.random.rand(100, 1)
y = 2 + 3 * x + np.random.randn(100, 1)
```

```python
x
```

```python
y
```

```python
D_model = keras.Sequential(layers.Dense(units = 1, input_dim=1, activation = 'linear'))#input_dim = 1 is model working on single variavle
```

```python
D_model.compile(optimizer='sgd', loss='mean_squared_error') #sgd = stochastic gradient discriminent
```

```python
D_model.fit(x, y, epochs=100, batch_size=32) #epochs is number of iterations and batchsize refers to no.of samples considered in one iteration
```

```python
new_data = np.array([[3.0], [3.0]])#giving a sample input to generate predictions
predictions = D_model.predict(new_data)
print("Predictions:")
print(predictions)
```

**On dataset**

```python
dataset = pd.read_csv('/content/Salary_Data.csv')
data = dataset.copy()
```

```python
x = data['YearsExperience']
y = data['Salary']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=32)
```

```python
model = keras.Sequential([layers.Dense(units=1,input_dim=1, activation='linear')])
model.compile(optimizer='sgd', loss ='mean_squared_error')
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=32)
```

```python
predict = model.predict(x_test)
```

```python
predictions = predict.reshape(-1,1)
predictions
```

```python
actual_vals = np.array(y_test)
actual_vals
```

```python
result = pd.DataFrame(predictions, columns=['predicted'])
result['actual'] = actual_vals
result
```
